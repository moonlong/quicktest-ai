```python
print(novel_test)
```

```python
def transform_func(inputs: dict) -> dict:
    text = inputs["text"]
    short_text = "\n\n".join(text.split("\n")[:10])
    return {"output_text": short_text}
```

```python
t_chain = TransformChain(
    input_variables=["text"],
    output_variables=["output_text"],
    transform=transform_func
)
```

```python
t_novel = t_chain(novel_test)
```

```python
print(t_novel)
```

```python
template = """总结下面文本:

{output_text}

总结:"""
```

```python
p = PromptTemplate(input_variables=["output_text"],outpu template=template)
```

```python
llm_chain = LLMChain(llm=OpenAI(temperature=0,max_tokens=1000),prompt=p,output_key="review",verbose=True)
```

```python
sequential_chain = SimpleSequentialChain(chains=[t_chain, llm_chain])
```

```python
sequential_chain.run(novel_test)
```

```python
ss_chain = SequentialChain(
    chains=[t_chain,llm_chain],
    input_variables=["text"],
    output_variables=["output_text","review"],
    verbose=True
)
```

```python
ss_chain(novel_test)
```
